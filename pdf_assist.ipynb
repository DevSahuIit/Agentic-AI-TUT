# PDF RAG Assistant using Agno + PgVector + Groq

## Objective

The objective of this project is to build an intelligent **Retrieval-Augmented Generation (RAG)** assistant that can answer questions from a PDF document. Instead of relying only on the Large Language Model's pre-trained knowledge, the system retrieves relevant information from the uploaded PDF and uses it to generate accurate, context-aware responses.

---

# Project Components

### 1. Agno

Agno is an open-source framework for building AI agents. It manages the interaction between the user, the knowledge base, the vector database, and the Large Language Model (LLM).

**Role in this project:**
- Creates the AI agent
- Connects the knowledge base with the LLM
- Retrieves relevant information from the vector database
- Maintains conversation history
- Generates responses for user queries

---

### 2. Groq LLM

Groq provides high-speed inference for open-source Large Language Models such as Llama 3.

**Role in this project:**
- Understands user questions
- Generates natural language responses
- Uses retrieved PDF content to produce accurate answers

---

### 3. PDF Knowledge Base

The PDF Knowledge Base is responsible for processing the PDF document before it can be searched.

Its tasks include:
- Downloading or reading the PDF
- Extracting text
- Splitting the document into smaller chunks
- Sending chunks for embedding generation
- Storing embeddings inside the vector database

---

### 4. Vector Database (PgVector)

A vector database stores embeddings instead of plain text.

An embedding is a numerical representation of text that captures its semantic meaning.

**Role in this project:**
- Stores vector embeddings
- Performs semantic similarity search
- Retrieves the most relevant document chunks for a query

---

### 5. PostgreSQL Storage

Besides storing embeddings, PostgreSQL is also used to save agent-related information.

It stores:
- Chat history
- User IDs
- Run IDs
- Previous conversations

This enables the assistant to continue conversations without losing context.

---

# Project Workflow

## Step 1: Load Environment Variables

The project first loads environment variables from a `.env` file.

These include:
- Groq API Key
- Database credentials

This approach improves security by keeping sensitive information outside the source code.

---

## Step 2: Connect to PostgreSQL

The application establishes a connection with the PostgreSQL database running inside a Docker container.

This database contains:
- Vector embeddings
- Chat history
- Agent storage

---

## Step 3: Create the Knowledge Base

The PDF is loaded into the Knowledge Base.

The Knowledge Base:
- Reads the document
- Extracts text
- Prepares it for embedding generation

---

## Step 4: Chunking

The extracted document is divided into smaller pieces called **chunks**.

Chunking improves retrieval accuracy because searching smaller sections is more effective than searching the entire document.

---

## Step 5: Embedding Generation

Each chunk is converted into a high-dimensional numerical vector using an embedding model.

Embeddings capture the semantic meaning of the text rather than exact keywords.

---

## Step 6: Store Embeddings

The generated embeddings are stored inside the PgVector database.

Once stored, they can be reused without regenerating them every time the application starts.

---

## Step 7: Create the AI Agent

The Agno Agent is initialized with:

- Knowledge Base
- PostgreSQL Storage
- Vector Database
- Groq LLM

The agent acts as the central controller of the entire RAG system.

---

## Step 8: User Query

The user asks a question in natural language.

Example:

> "How do I prepare chocolate cake?"

---

## Step 9: Similarity Search

Instead of searching by keywords, the system converts the user query into an embedding.

This embedding is compared with all stored document embeddings.

The most semantically similar chunks are retrieved.

---

## Step 10: Response Generation

The retrieved document chunks are sent to the Groq LLM along with the user's question.

The LLM uses this additional context to generate a factual and context-aware response.

---

## Step 11: Store Conversation

The conversation is saved inside PostgreSQL.

This allows:
- Multi-turn conversations
- Memory across sessions
- Personalized interactions

---

# Why Use a Vector Database?

Traditional databases search for exact words.

Vector databases search based on **meaning**.

For example:

Question:

> "How can I bake a cake?"

Stored text:

> "Recipe for preparing chocolate cake."

Even though the wording is different, the vector database recognizes that both texts have similar meanings and retrieves the correct information.

---

# Why Docker is Used

Docker is used to run the PostgreSQL + PgVector database inside an isolated container.

Benefits include:

- Easy setup
- No manual installation
- Platform independence
- Consistent environment across systems
- Simplified deployment

- Scalable architecture
- Faster inference using Groq
- Easy deployment with Docker

# Overall Workflow

```
                 User
                  │
                  ▼
           User Question
                  │
                  ▼
             Agno Agent
                  │
                  ▼
        Convert Query to Embedding
                  │
                  ▼
       PgVector Similarity Search
                  │
                  ▼
      Retrieve Relevant PDF Chunks
                  │
                  ▼
          Send Context to Groq LLM
                  │
                  ▼
        Generate Contextual Answer
                  │
                  ▼
          Return Response to User
                  │
                  ▼
      Store Chat History in PostgreSQL
```

By combining semantic search with a Large Language Model, the assistant can provide accurate, context-aware answers from PDF documents while maintaining conversation history for an improved user experience.
