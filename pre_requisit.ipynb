# Agentic AI

## Definition
**Agentic AI** is an AI system that can autonomously **plan, reason, make decisions, and use tools** to accomplish a goal with minimal human intervention.

## How Agentic AI Works

1. **Receives a Goal** – Understands the user's objective.
2. **Plans** – Breaks the goal into smaller, manageable tasks.
3. **Reasons** – Determines the best strategy for completing each task.
4. **Uses Tools** – Interacts with APIs, databases, web search, or other software when required.
5. **Executes Actions** – Performs the planned tasks step by step.
6. **Monitors Progress** – Evaluates outcomes and adjusts its approach if needed.
7. **Delivers the Result** – Completes the objective and provides the final output to the user.

# Difference Between Generative AI, AI Agents, and Agentic AI

| Feature | Generative AI | AI Agent | Agentic AI |
|---------|---------------|-----------|------------|
| **Definition** | AI that generates content such as text, images, code, or audio based on user prompts. | An AI system that performs specific tasks by interacting with tools or environments. | An advanced AI system that autonomously plans, reasons, and executes multiple tasks to achieve a goal. |
| **Primary Purpose** | Generate content | Perform assigned tasks | Achieve goals autonomously |
| **Decision Making** | Limited | Rule-based or predefined | Dynamic and autonomous |
| **Planning** | No | Basic | Multi-step planning |
| **Reasoning** | Limited | Moderate | Advanced |
| **Tool Usage** | Usually none (unless integrated) | Uses tools required for a task | Selects and uses multiple tools intelligently |
| **Human Intervention** | Required for every prompt | Required for task initiation | Minimal; only for approvals or guidance |
| **Memory** | Limited conversation context | May store task state | Maintains long-term context and progress |
| **Autonomy** | Low | Medium | High |
| **Example** | Writing an essay with ChatGPT | A chatbot that books movie tickets | A virtual assistant that plans a trip, books flights and hotels, schedules meetings, and updates your calendar automatically |

## Summary

- **Generative AI** → Creates content from prompts.
- **AI Agent** → Performs specific tasks using tools.
- **Agentic AI** → Plans, reasons, makes decisions, and autonomously executes multiple tasks to achieve a goal.

# Docker for Vector Databases

## What is Docker?

**Docker** is a platform that allows you to package an application and all of its dependencies into a lightweight, portable container. A container ensures that the application runs the same way on any system without worrying about software installation or version conflicts.

**In simple words:**
> Docker is like a virtual box that contains your application along with everything it needs to run.

---

# Why Do We Use Docker?

Docker is used to:

- Avoid installing software directly on your computer.
- Keep applications isolated from each other.
- Ensure the same environment across different machines.
- Easily deploy applications.
- Run databases and servers with a single command.

---

# Why is Docker Used for Vector Databases?

Most vector databases (such as **PgVector, Qdrant, Weaviate, Milvus, Chroma Server**) require:

- A database server
- Storage management
- Networking
- Configuration files
- Background services

Installing these manually is difficult.

Docker packages everything together, allowing the vector database to run in an isolated environment without affecting your system.

# Running PgVector using Docker

Run:

```bash
docker run -d \
--name pgvector \
-e POSTGRES_USER=ai \
-e POSTGRES_PASSWORD=ai123 \
-e POSTGRES_DB=vectordb \
-p 5532:5432 \
pgvector/pgvector:pg16
```

This command:

- Downloads the PgVector image (if needed)
- Creates a PostgreSQL database
- Enables the pgvector extension
- Opens port **5532**
- Starts the database
